```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
weather = pd.read_csv(r'weather.csv', index_col=0) # index_col=0:第一列为索引值
```

```python
weather.head()
```

```python
# 将特征矩阵和标签分开
X = weather.iloc[:, :-1]
Y = weather.iloc[:, -1]
```

```python
X.shape
```

```python
Y.shape
```

```python
X.info()
```

```python
X.isnull().mean() # 缺失值
```

```python
Y.isnull().mean()
```

```python
np.unique(Y) # 探索标签的分类
```

```python
# 将数据分为训练集和测试集
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.3, random_state=420)
for i in [Xtrain, Xtest, Ytrain, Ytest]: # 恢复索引
    i.index = range(i.shape[0])
```

```python
# 样本不平衡问题探索
Ytrain.value_counts()
```

```python
Ytest.value_counts()
```

```python
# 将标签进行编码
from sklearn.preprocessing import LabelEncoder
encorder = LabelEncoder().fit(Ytrain) 
Ytrain = pd.DataFrame(encorder.transform(Ytrain))
Ytest = pd.DataFrame(encorder.transform(Ytest))
```
