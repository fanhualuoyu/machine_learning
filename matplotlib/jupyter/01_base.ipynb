```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# 设置绘图样式
plt.style.use('classic')
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.show()
```

```python
fig.savefig('base.png')
```

```python
# 两种画图接口
# MATLAB风格接口
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
# 面向对象接口
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
# 简易线形图
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
# 线条的颜色与风格
plt.plot(x, np.sin(x - 0), color='blue') # 标准颜色名称 
plt.plot(x, np.sin(x - 1), color='g') # 缩写颜色代码（rgbcmyk） 
plt.plot(x, np.sin(x - 2), color='0.75') # 范围在0~1的灰度值 
plt.plot(x, np.sin(x - 3), color='#FFDD44') # 十六进制（RRGGBB，00~FF） 
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB元组，范围在0~1 
plt.plot(x, np.sin(x - 5), color='chartreuse'); # HTML颜色名称
```

```python
# 线条的风格
plt.plot(x, x + 0, linestyle='solid') 
plt.plot(x, x + 1, linestyle='dashed') 
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
```

```python
# 简化形式
plt.plot(x, x + 4, linestyle='-') # 实线 
plt.plot(x, x + 5, linestyle='--') # 虚线 
plt.plot(x, x + 6, linestyle='-.') # 点划线 
plt.plot(x, x + 7, linestyle=':'); # 实点线
```

```python
# 坐标轴上下限
plt.plot(x, np.sin(x)) 
plt.xlim(-1, 11) 
plt.ylim(-1.5, 1.5);
```

```python
# 坐标轴逆序显示
plt.plot(x, np.sin(x)) 
plt.xlim(10, 0) 
plt.ylim(1.2, -1.2);
```

```python
plt.plot(x, np.sin(x)) 
plt.axis([-1, 11, -1.5, 1.5]);# plt.axis()（注意不要搞混 axes 和 axis）。通过传入 [xmin, xmax, ymin, ymax]
```

```python
# 不留空白区域
plt.plot(x, np.sin(x)) 
plt.axis('tight');
```

```python
# 收紧布局示例
plt.plot(x, np.sin(x))
plt.axis('equal')
```

```python
# 设置图形标签
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
# 图形标题与坐标轴标题
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend()
```

```python
ax = plt.axes() 
ax.plot(x, np.sin(x)) 
ax.set(xlim=(0, 10), ylim=(-2, 2),
        xlabel='x', ylabel='sin(x)', title='A Simple Plot');
```
