```python
# 直方图
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

```python
data = np.random.randn(1000)
plt.hist(data)
```

```python
plt.hist(data, bins=30, normed=True, alpha=0.5, histtype='stepfilled', color='steelblue', edgcolor='none')
```

```python
# 同坐标轴的多个频次直方图
# histtype='stepfilled' 与透明性设置参数 alpha 搭配
x1 = np.random.normal(0, 0.8, 1000) 
x2 = np.random.normal(-2, 1, 1000) 
x3 = np.random.normal(3, 2, 1000) 
kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40) 
plt.hist(x1, **kwargs) 
plt.hist(x2, **kwargs) 
plt.hist(x3, **kwargs)
```

```python
# 计算频次直方图（就是计算每段区间的样本数）
counts, bin_edges = np.histogram(data, bins=5)
```

```python
# 二维频次直方图
mean = [0, 0] 
cov = [[1, 1], [1, 2]] 
x, y = np.random.multivariate_normal(mean, cov, 10000).T # 多元高斯分析生成x与y的样本数据
```

```python
x
```

```python
y
```

```python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
counts, xedges, yedges = np.histogram2d(x, y, bins=30)
```

```python
# plt.hexbin：六边形区间划分
plt.hexbin(x, y, gridsize=30, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```
